###**1. Crear un documento Google Colaboratory y cargar los set de datos como DataFrames.**

````python
#Importar los librerias
import pandas as pd
import numpy as np
#Cargar los Dataframes con los set de datos
df_ventas=pd.read_csv('ventas.csv')
df_clientes=pd.read_csv('clientes.csv')
df_marketing=pd.read_csv('marketing.csv')
````

###**2. Realizar un script básico que calcule las ventas mensuales utilizando variables y operadores.**

````python
# Abrir el archivo CSV
with open('ventas.csv', 'r', encoding='utf-8') as archivo:
    lineas = archivo.readlines()

# Quitar el encabezado
lineas = lineas[1:]

# Diccionario para acumular las ventas por mes
ventas_mensuales = {}

# Recorrer cada línea
for linea in lineas:
    linea = linea.strip()
    if not linea:  # Saltar líneas vacías
        continue

    partes = linea.split(',')

    # Asegurar que tenga al menos 6 columnas
    if len(partes) < 6:
        continue

    id_venta = partes[0].strip()
    producto = partes[1].strip()
    precio_str = partes[2].strip()
    cantidad_str = partes[3].strip()
    fecha_venta = partes[4].strip()
    categoria = partes[5].strip()

    # Limpiar y validar precio
    precio_str = precio_str.replace('$', '')
    #.replace(' ', '').replace('.', '').replace(',', '.')
    if not precio_str:
        continue  # si está vacío, salteamos
    try:
        precio = float(precio_str)
    except ValueError:
        continue  # si no se puede convertir, se salta

    # Validar cantidad
    if not cantidad_str.isdigit():
        continue
    cantidad = int(cantidad_str)

    # Calcular total
    total_venta = precio * cantidad

    # Obtener mes de la fecha (DD/MM/YYYY)
    if len(fecha_venta) >= 5:
        mes = fecha_venta[3:5]
    else:
        continue  # si la fecha no tiene formato válido

    # Acumular
    if mes in ventas_mensuales:
        ventas_mensuales[mes] += total_venta
    else:
        ventas_mensuales[mes] = total_venta

# Mostrar resultados
total_general=0
print("=== Ventas mensuales ===")
for mes, total in ventas_mensuales.items():
    print(f"Mes {mes}: ${total:,.2f}")
    total_general=total_general+total
print(f"Total general de ventas: ${total_general:,.2f}")
````

````python
# Estructuras de datos: Lista de diccionarios (Me parece que genera un codigo mas legible)

ventas = []  # lista vacía para almacenar las ventas

# Abrir el archivo CSV
with open('ventas.csv', 'r', encoding='utf-8') as archivo:
    lineas = archivo.readlines()

# Quitar el encabezado
lineas = lineas[1:]

# Diccionario para acumular las ventas por mes
ventas_mensuales = {}

# Recorrer cada línea
for linea in lineas:
    linea = linea.strip()
    if not linea:  # Saltar líneas vacías
        continue

    partes = linea.split(',')

    # Asegurar que tenga al menos 6 columnas
    if len(partes) < 6:
        continue

    id_venta = partes[0].strip()
    producto = partes[1].strip()
    precio_str = partes[2].strip()
    cantidad_str = partes[3].strip()
    fecha_venta = partes[4].strip()
    categoria = partes[5].strip()

    # Limpiar y validar precio
    precio_str = precio_str.replace('$', '').replace(' ', '').replace('.', '').replace(',', '.')
    if not precio_str:
        continue  # si está vacío, salteamos
    try:
        precio = float(precio_str)
    except ValueError:
        continue  # si no se puede convertir, se salta

    # Validar cantidad
    if not cantidad_str.isdigit():
        continue
    cantidad = int(cantidad_str)

    # Calcular total
    total_venta = precio * cantidad

    # Obtener mes de la fecha (DD/MM/YYYY)
    if len(fecha_venta) >= 5:
        mes = fecha_venta[3:5]
    else:
        continue  # si la fecha no tiene formato válido

    # Acumular
    if mes in ventas_mensuales:
        ventas_mensuales[mes] += total_venta
    else:
        ventas_mensuales[mes] = total_venta
 # Crear el diccionario para esta venta
    venta = {
        "producto": producto,
        "precio": precio,
        "cantidad": cantidad
    }

    # Agregarlo a la lista
    ventas.append(venta)

# Mostrar todas las ventas almacenadas
total_general=0
print("=== Ventas registradas ===")
cont=0
for venta in ventas:
    total = venta["precio"] * venta["cantidad"]
    #condicion para imprimir solo 5, por se sino es muy largo el listado
    if cont<5:
      print(f"Producto: {venta['producto']}, Precio: ${venta['precio']:,.2f}, Cantidad: {venta['cantidad']}, Total: ${total:,.2f}")
      cont+=1
    # Calcular el total general
    total_general += total

# Mostrar el total general
print(f"\nTotal general de ventas: ${total_general:,.2f}")
````

##**Etapa 2: Preprocesamiento y Limpieza de Datos (Clases 5 a 8)**

###**1- Limpieza de Datos: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados. Documentar el proceso y los resultados.**

####Quitar Nulos

````python
#Quitar elementos Nulos de Ventas

#Guardar las filas con valores nulos
df_ventas_nulos = df_ventas[df_ventas.isna().any(axis=1)]

#Eliminar las filas nulas del DataFrame original
df_ventas = df_ventas.dropna()

print("Registros Eliminados por tener Nulos")
display(df_ventas_nulos)
````

**Registros Eliminados por tener Nulos**
```
id_venta	producto	precio	cantidad	fecha_venta	categoria
136	627	Elementos de cerámica	NaN	NaN	17/01/2024	Decoración
139	2171	Parlantes Bluetooth	NaN	NaN	17/01/2024	Electrónica
```

####Quitar caracteres no deseados y convertir a lo tipos correctos

````python
#Quitar caracteres no deseados del precio y convertir a tipo float
df_ventas['precio'] = df_ventas['precio'].replace(r'[\$,]', '', regex=True).astype(float)

#Quitar espacios al inicio y al final de producto y categoria
df_ventas['producto'] = df_ventas['producto'].str.strip()
df_ventas['categoria'] = df_ventas['categoria'].str.strip()

#Convertir fecha_venta a tipo datetime
df_ventas['fecha_venta'] = pd.to_datetime(df_ventas['fecha_venta'], dayfirst=True, errors='coerce')
df_marketing['fecha_inicio'] = pd.to_datetime(df_marketing['fecha_inicio'], dayfirst=True, errors='coerce')
df_marketing['fecha_fin'] = pd.to_datetime(df_marketing['fecha_fin'], dayfirst=True, errors='coerce')

#Mostrar cantidad de registros sin convertir y como quedo df_ventas con los cambios de tipo de dato.
print(f"Registros de Ventas sin poder convertirse a datetime = {df_ventas['fecha_venta'].isna().sum()}")
print(f"Registros de Marketing (fecha_inicio) sin poder convertirse a datetime = {df_marketing['fecha_inicio'].isna().sum()}")
print(f"Registros de Marketing (fecha_fin) sin poder convertirse a datetime = {df_marketing['fecha_fin'].isna().sum()}")

display(df_ventas.info())
display(df_marketing.info())
````

####Quitar Duplicados

````python
#Mostrar elementos a eliminar por se duplicado
print(f"Total de duplicados eliminados = {df_ventas.duplicated(keep='first').sum()}")

#Eliminar los duplicados del DataFrame original
df_ventas = df_ventas.drop_duplicates(keep="first")

#Mostrar el DataFrame sin duplicados
print("DataFrame sin duplicados")
display(df_ventas)
````

###**2- Transformación de Datos: Aplicar filtros y transformaciones para crear una tabla de ventas que muestre solo los productos con alto rendimiento.**

````python
#Crear una columna con el total de cada venta
df_ventas['total_venta'] = df_ventas['precio'] * df_ventas['cantidad']

#Agrupar por producto y calcular métricas
df_rendimiento = df_ventas.groupby('producto', as_index=False).agg({
    'cantidad': 'sum',
    'total_venta': 'sum'
})

#Calcular el promedio de ventas para definir "alto rendimiento"
promedio_ventas = df_rendimiento['total_venta'].mean()

#Filtrar productos que superen el promedio
df_alto_rendimiento = df_rendimiento[df_rendimiento['total_venta'] > promedio_ventas]

# Ordenar de mayor a menor
df_alto_rendimiento = df_alto_rendimiento.sort_values(by='total_venta', ascending=False)

#Mostrar resultados
print("=== Productos de alto rendimiento ===")
print(f"Promedio de ventas es = {promedio_ventas:,.2f}")

display(df_alto_rendimiento)

#Mostrar total de ingreso de lo productos con mejores rendimientos
print(f"Total de ingresos de productos con mejores rendimientos= ${df_alto_rendimiento['total_venta'].sum():,.2f}")
````

###**3- Agregación: Resumir las ventas por categoría de producto y analizar los ingresos generados.**

````python
#Agrupar por categoría y sumar cantidad e ingresos
df_categoria = df_ventas.groupby('categoria', as_index=False).agg({
    'cantidad': 'sum',       # total de unidades vendidas por categoría
    'total_venta': 'sum'     # ingresos totales por categoría
})

#Ordenar por ingresos descendente
df_categoria = df_categoria.sort_values(by='total_venta', ascending=False)

#Mostrar resultados
print("=== Ventas resumidas por categoría ===")
display(df_categoria)

# Mostar Total de Ingreso
print(f"Total de ingresos = ${df_categoria['total_venta'].sum():,.2f}")
````

###**4- Integración de Datos: Combinar los sets de datos de ventas y marketing para obtener una visión más amplia de las tendencias.**

````python
df_ventas.head()
#Agrupo df_ventas por producto
df_ventas_agrupadas = df_ventas.groupby('producto', as_index=False).agg({
    'cantidad': 'sum',
    'total_venta': 'sum'
})
#Agrupo df_marketing por producto
df_marketing_agrupado = df_marketing.groupby('producto', as_index=False).agg({
    'costo': 'sum',
    'id_campanha':'count'
})

#integrar los 2 df
df_integracion = pd.merge(df_ventas_agrupadas, df_marketing_agrupado, on='producto', how='outer')
df_integracion.sort_values(by='total_venta', ascending=False, inplace=True)

df_integracion.head()
````
